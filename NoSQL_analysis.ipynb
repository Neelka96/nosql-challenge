### 1. Which establishments have a hygiene score equal to 20?

```python
from collections import defaultdict
params = defaultdict(dict)

# Function to complete 3 standard tasks
def explore_db(collection, params):
    '''Takes in <pymongo collection object> to perform 3 standard analysis tasks'''
    query = params.get('query', None)
    fields = params.get('fields', None)
    sort = params.get('sort', None)
    limit = params.get('limit', None)

    if fields:
        results = collection.find(query, fields)
    else:
        results = collection.find(query)

    if sort:
        results = results.sort(sort)
    if limit:
        results = results.limit(limit)

    # Display the number of documents in the result (originally done using `.count_documents(query)`)
    results = [*results]
    
    count = len(results)
    print(f'Document Count: {count}\n\n')
    
    # Display the first document in the results using pprint (originally done using `.find_one()`)
    pprint(results[0])
    print('\n\n\n')

    df = pd.DataFrame(results)
    print(f'# of Rows in DF: {df.shape[0]}\n\n')
    df.head(10)

    return df
```

```python
# For the establishments with a hygiene score of 20
params['query'] = {'scores.Hygiene': 20}
explore_db(establishments, params)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
params['query'] = {
    'LocalAuthorityName': {'$regex': 'London'}
    ,'RatingValue': {'$gte': 4}
}

explore_db(establishments, params)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
query = {'BusinessName': 'Penang Flavours'}
fields = {'geocode.latitude': 1, 'geocode.longitude': 1}
result = [*establishments.find(query, fields)][0]

degree_search = 0.01
latitude = result['geocode']['latitude']
latitude = [latitude - degree_search, latitude + degree_search]

longitude = result['geocode']['longitude']
longitude = [longitude - degree_search, longitude + degree_search]


params['query'] = {
    'RatingValue': 5
    ,'geocode.latitude': {
        '$gte': latitude[0]
        ,'$lte': latitude[1]
    }
    ,'geocode.longitude': {
        '$gte': longitude[0]
        ,'$lte': longitude[1]
    }
}
params['sort'] = [('scores.Hygiene', 1)]
params['limit'] = 5

explore_db(establishments, params)
```
